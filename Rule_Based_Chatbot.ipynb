this is chatbot named Gie, used for helping Rookies who have any question about Summer Club by Generation Girl



```python
import re
import random

# Dictionary of Gie's responses with lemmatized keywords
responses = {
    "hello": {
        "patterns": [r'\bhi\b|\bhello\b|\bhey\b'],
        "responses": ["Hello Rookies! Welcome to Summer Club Explorer 2023! How can Gie assist you?"]
    },
    "menu": {
        "patterns": [r'\b[mM][wnr]*u\b'],
        "responses": ["Certainly! Here's what Gie can help you with:\n1. Schedule information\n2. Preparation before classes\n3. Office Hour and guidance\n4. Rules and requirements\n5. Emergency questions\n\nPlease choose one by number or ask your question."]
    },
    "difficulty": {
        "patterns": [r'\bdifficulty\b|\bconfus.*\b|\bfail+\b|\bdon\'t understand\b|\bhelp\b|\bassist\b'],
        "responses": ["If you're facing difficulties, don't hesitate to ask a mentor or TA during Office Hour. We are here to help you overcome any challenges."]
    },
    "to-do list": {
        "patterns": [r'\b(to-do list|tasks|list)\b'],
        "responses": [
            "Here's your to-do list:\n"
            "- [ ] [Join Whatsapp Group Community](https://chat.whatsapp.com/DwBN0iydjW2KQLxbrioF78)\n"
            "- [ ] Download **[Zoom Background](https://drive.google.com/file/d/1eEocHlPf5BkhUdICRNvjD0Gw6tHw5ZOt/view?usp=sharing)**\n"
            "- [ ] [Link to Zoom Meetings](https://bit.ly/summerclubexplorer-2023)\n"
            "- [ ] [**Sign up on Thinkific**](http://generationgirl.thinkific.com) and complete the **[Class Prep Forms](https://generationgirl.typeform.com/to/rYsYZVez)**\nNote: You can access study materials before classes on Thinkific!"
        ]
    },
    "schedule": {
        "patterns": [r'\bschedule\b|\btimetab.*\b|\b1\b'],
        "responses": ["You can find the program schedule [here](https://www.notion.so/Summer-Club-Explorer-2023-Rookies-Handbook-0634ac131af44db08a62100e6b8434a1?pvs=21).", "You can check the complete program schedule in our handbook."]
    },
    "preparation": {
        "patterns": [r'\bpreparation\b|\bstudy\b|\bhow to study\b|\b2\b'],
        "responses": ["To prepare effectively, make sure you've registered on Thinkific and accessed the study materials. It's important to study before attending classes."]
    },
    "office_hour": {
        "patterns": [r'\boffic.*\b|\bguidance\b|\bmentor\b|\bTA\b|\b3\b'],
        "responses": ["If you need guidance, you can schedule an Office Hour with a mentor or TA outside of class hours. Here's how to book an Office Hour:\n1. Join the [Whatsapp Group Community](https://www.notion.so/Summer-Club-Explorer-2023-Rookies-Handbook-0634ac131af44db08a62100e6b8434a1?pvs=21)\n2. Arrange a meeting with the mentor you want to have an Office Hour with\n3. Attend the Google Meet scheduled by the mentor"]
    },
    "whatsapp_group": {
        "patterns": [r'\bwhatsapp\b|\bwhatsapp group\b'],
        "responses": [
            "You can join the WhatsApp groups based on your chosen topic:\n"
            "- Group General (For general information), you MUST join this group [here](https://chat.whatsapp.com/DwBN0iydjW2KQLxbrioF78)\n"
            "- UI/UX Week 1 Class [here](https://chat.whatsapp.com/JD8vZOII9Dv59xkU9jkBtK)\n"
            "- Computer Science Class [here](https://chat.whatsapp.com/C0GtB4aZbR7InJR3dxnsz9)\n"
            "- Build A Website (FE) Class [here](https://chat.whatsapp.com/F2sG5p8sxLIHzSg3vWA2sW)\n"
            "- UI/UX Week 2 Class [here](https://chat.whatsapp.com/Ly5bztynxaX1FJKs7Ac1yx)\n"
            "- Data Science Class [here](https://chat.whatsapp.com/LFJu4Z7D5YeKy177ppU8dx)\n"
            "- Build A Website (BE) Class [here](https://chat.whatsapp.com/JWK118VV4CR27q6rs7ycuQ)"
        ]
    },
    "rules": {
        "patterns": [r'\brules\b|\bguidelines\b|\brequirements\b|\b4\b'],
        "responses": [
            "⚠️ Rules\n\n"
            "- Classes will start at 19:00 every day. Please make sure to be on time, with your camera and microphone on! We want to see and hear you🥰\n"
            "- You must attend the class sessions every day. However, if you can't attend a class, make sure to inform the Mentor/TA!\n"
            "- You must complete the Final Project and fill out the [Important Form](https://www.notion.so/Summer-Club-Explorer-2023-Rookies-Handbook-0634ac131af44db08a62100e6b8434a1?pvs=21) to receive a certificate and merchandise.\n"
            "- E-Certificate and merchandise will be given if participants meet all the requirements below:\n"
            "    - Minimum 80% attendance\n"
            "    - Complete the pre and post-tests, as specified in the 'Overall Program Agenda' above\n"
            "    - Submit the Final Project\n\n"
            "If any of these requirements are not met, the E-Certificate and Merchandise cannot be sent to participants!"
        ]
    },
    "help": {
        "patterns": [r'\bhelp\b|\bassist\b|\bhelp me\b|\bhelp with something\b'],
        "responses": ["Certainly, feel free to ask any questions you have!"]
    },
    "exit": {
        "patterns": [r'\bexit\b|\bquit\b|\bstop\b'],
        "responses": ["Goodbye! If you have more questions in the future, don't hesitate to ask."]
    },
}

# Function to respond to user input
def respond(user_input):
    user_input = user_input.lower()

    # Periksa apakah kata "menu", "mwnu", atau "mrnu" ada dalam input dan anggap sebagai "menu"
    if re.search(r'[mM][enu\.]+', user_input):
        return responses["menu"]["responses"][0]

    for intent, data in responses.items():
        for pattern in data["patterns"]:
            if re.search(pattern, user_input):
                if intent == "rules":
                    return responses["rules"]["responses"][int(user_input) - 1]
                else:
                    return random.choice(data["responses"])

    if re.match(r'\b[1-5]\b', user_input):
        return "Please choose one by number or ask your question."
    else:
        return "Only numbers corresponding to the menu options (1-5) are accepted. Please choose one of those or ask your question."

# Main program
first_interaction = True
print("👋 Hello Rookies!")
print("Welcome to the Rookies Handbook 🎉")
print("How can Gie assist you?")
print("Type 'menu' to see available options or 'exit' to quit.")

while True:
    user_input = input("Rookies: ")
    if user_input.lower() == 'exit':
        print("Thank you, Rookies, for using this chatbot. Happy learning!")
        break
    else:
        response = respond(user_input)
        print("Gie:", response)
```
